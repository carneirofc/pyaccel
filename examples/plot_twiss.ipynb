# Plot of Twiss parameters

#### Loads libraries and setting environment

```python
from pyaccel.interactive import *
%matplotlib inline
```

#### Creates SIRIUS storage ring model (last version) and turning cavity and radiation on for 6D trackings

```python
the_ring = si_model.create_accelerator()
set6dtracking(the_ring)
```

#### Calculates Twiss parameters

```python
twiss = calctwiss(the_ring)
spos, betax, betay, etax = gettwiss(twiss, ('spos', 'betax','betay', 'etax'))
print('Twiss at start of the Sirius lattice:')
print(twiss[0])
```

#### Plots Twiss parameters

```python
plt.plot(spos, betax)
plt.plot(spos, betay)
plt.plot(spos, 100*etax)
circumference = the_ring.length
symmetry = si_model.lattice_symmetry
_ = plt.xlim((0,circumference/symmetry))
_ = plt.xlabel('spos [m]')
_ = plt.ylabel('betax [m], betay [m], etax [cm]')
_ = plt.legend(('betax','betay', 'etax'))
_ = plt.grid('on')
_ = plt.title('Twiss Functions for SIRIUS SI_V07.C05')
```
